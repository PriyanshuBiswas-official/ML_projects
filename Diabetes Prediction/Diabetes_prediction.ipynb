# **Adding pipeline for preprocessing**

```python
#Data Standardization
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
my_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])
df_tr = my_pipeline.fit_transform(x)
```

```python
df_tr
```

```python
x = df_tr
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state = 2)
```

```python
x.shape,x_train.shape, x_test.shape
```

```python
y.shape,y_train.shape,y_test.shape
```

# **Training the model**

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
model = svm.SVC(kernel = 'linear')
```

```python
# fitting the model to the dataset
model.fit(x_train,y_train)
```

```python
x_train_pred = model.predict(x_train)
```

```python
x_train_pred
```

# **Testing accuracy for both x_train and x_test**



```python
accuracy_x_train = accuracy_score(x_train_pred, y_train)
```

```python
print("The accuracy score of the training data: ",accuracy_x_train)
```

```python
x_test_pred = model.predict(x_test)
x_test_pred
```

```python
accuracy_x_test = accuracy_score(x_test_pred, y_test)
```

```python
print("The accuracy score of the test data: ",accuracy_x_test)
```
