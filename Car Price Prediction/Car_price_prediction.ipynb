```python
#editing categorical data

df['Fuel'] = df['Fuel'].map({
    'Petrol': 0,
    'Diesel': 1,
    'CNG': 2,
    'LPG': 3,
    'Electric': 4
})

df['Transmission'] = df['Transmission'].map({
    'Manual': 0,
    'Automatic': 1
})

df['Seller_Type'] = df['Seller_Type'].map({
    'Individual': 0,
    'Dealer': 1,
    'Trustmark Dealer': 2
})
df['Owner'] = df['Owner'].map({
    'First Owner': 0,
    'Second Owner': 1,
    'Third Owner': 2,
    'Fourth & Above Owner':3,
    'Test Drive Car':4
})
```

```python
# encoding categorical data into the following 
df = pd.get_dummies(df, columns=['Brand', 'Model'], drop_first=True)
```

```python
df.head()
```

```python
# separating data
x = df.drop(['Selling_Price'], axis = 1)
y = df['Selling_Price']
```

```python
print(x)
print(y)
```

```python
# splitting data into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 2)
```

```python
# loading linear regression model from sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# fitting the model to the training data
model.fit(x_train, y_train)
```

```python
# checking cross validation scores before predicting
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x, y, cv=5, scoring='r2')
print(scores)
print("Mean CV R2:", scores.mean())
```

```python
# predicting using training data
model_pred = model.predict(x_train)
```

```python
# R^2 score for training data
from sklearn.metrics import r2_score
error_score = r2_score(y_train, model_pred)
print("R squared error for Training data =",error_score)
```

```python
# Visualization of training actual vs predicted prices
plt.scatter(y_train, model_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Price")
plt.show()
```

```python
# R^2 score for test data
model_pred_test = model.predict(x_test)
error_train = r2_score(y_test, model_pred_test)
print("R squared error for Testing data =",error_score)
```

```python
# Visualization of testing actual vs predicted prices
plt.scatter(y_test, model_pred_test)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Price")
plt.show()
```
